# <center>Using Akaike Information Criterion to determine the best model</center>

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Assigning the data
data = np.loadtxt("data.dat") # Reading the file
x = data[0:,0] # Getting x data from the first col
y = data[0:,1] # Getting y data from the second col
```

The baseline to build the model line of a linear, quadratic and cubic model line are:

- Linear: y = $a_0$ + $a_1x$
- Quadratic: y = $a_0$ + $a_1x$ + $a_2x^2$
- Cubic: y = $a_0$ + $a_1x$ + $a_2x^2$ + $a_3x^3$

We use Normal Equation to find all the coefficients, by calculating the least-squares solution of matrix A (containing ones and the values of all x (including $x^2$ for quadratic, and $x^2$ and $x^2$ for cubic)) and vector y (values of y). We use np.linalg.listsq to compute the result, and get the first row where the coefficients values are located. Then, calculate the y-hat, which is the prediciton of the data (the model prediction), by multiplying matrix A with the coefficients. Finally we calculate the Sum of Squared Differences (SSE) to know the error between the model prediction and the data.



```python
# Linear model line
n = x.shape[0] # Get the size of array
A1 = np.ones((n,2)) # Make the matrix for linear model line
A1[:,1] = x # Insert x to the matrix
c1 = np.linalg.lstsq(A1, y, rcond=None)[0] # Returns the value of the coefficients

yhat1 = A1.dot(c1) # Finding the y hat
SSE1 = np.sum((y - yhat1)**2) # Finding the sum square error
```

```python
# Quadratic model line
A2 = np.ones((n,3)) # Make the matrix for quadratic model line
A2[:,1] = x # Insert x to the matrix
A2[:,2] = x**2 # Insert x**2 to the matrix
c2 = np.linalg.lstsq(A2, y, rcond=None)[0]

yhat2 = A2.dot(c2) 
SSE2 = np.sum((y - yhat2)**2)
```

```python
# Cubic model line
A3 = np.ones((n,4)) # Make the matrix for cubic model line
A3[:,1] = x # Insert x to the matrix
A3[:,2] = x**2 # Insert x**2 to the matrix
A3[:,3] = x**3 # Insert x**3 to the matrix
c3 = np.linalg.lstsq(A3, y, rcond=None)[0] 

yhat3 = A3.dot(c3) 
SSE3 = np.sum((y - yhat3)**2)
```

```python
# Quartic model line
A4 = np.ones((n,5)) # Make the matrix for cubic model line
A4[:,1] = x # Insert x to the matrix
A4[:,2] = x**2 # Insert x**2 to the matrix
A4[:,3] = x**3 # Insert x**3 to the matrix
A4[:,4] = x**4 # Insert x**3 to the matrix
c4 = np.linalg.lstsq(A4, y, rcond=None)[0] 

yhat4 = A4.dot(c4) 
SSE4 = np.sum((y - yhat4)**2)
```

```python
# Plotting the graph
plt.plot(x, y, 'mo', label='original data')
plt.plot(x, yhat1, label='linear model line')
plt.plot(x, yhat2, label='quadratic model line')
plt.plot(x, yhat3, label='cubic model line')
plt.plot(x, yhat4, label='quartic model line', color='dimgray')
plt.title('Graph to compare various model lines', fontweight='bold', fontsize='14')
plt.xlabel('x data')
plt.ylabel('y data')
plt.legend()
```

## <center>Comparing the Akaike Information Criterion (AIC)</center>
<center>The formula of AIC:</center>




Where K is the parameters, SSE is the sum of squared differences and n is the number of data
AIC is one of the best technique to calculate the best model because it uses a small SSE and simple model (parsimonious, meaning it uses minimum parameters

```python
SSE = np.array([SSE1, SSE2, SSE3, SSE4])
k = np.array([1, 2, 3, 4])
aic = n * np.log(SSE / n) + 2 * k

# Plotting graph k vs SSE
plt.bar(k, SSE, color='darkslateblue', edgecolor='black')
plt.title('Comparison between k (parameter) and SSE', fontweight='bold', fontsize='14')
plt.xlabel('k (parameter)')
plt.ylabel('SSE')
plt.xticks([1, 2, 3, 4])
```

```python
# Plotting graph k vs AIC
plt.bar(k, aic, color='mediumpurple', edgecolor='black')
plt.title('Comparison between k (parameter) and AIC', fontweight='bold', fontsize='14')
plt.xlabel('k (parameter)')
plt.ylabel('AIC')
plt.xticks([1, 2, 3, 4])
```

## Conclusion
From the graph we use to compare various model lines, we can see that the original data has a curve-like shape, so we can infer that either the quadratic model line, the cubic model line or the quartic model line would become the best model. The first model (linear) is not the best for this set of data. From the graph comparing the SSEs, the third model (cubic) has a slight difference from the second model (quadratic) and the fourth model (quartic) is the lowest. From the graph comparing AIC, we can conclude that the best model is the fourth model. However, we cannot conclude that the lower the SSE means the better model as we can see in model 2 and model 3. Although model 3 has a lower SSE, model 2's AIC is lower than model 3's, which means model 3 is worse than model 2. This means that we cannot conclude which model is the best based on the result of the lowest SSE alone. AIC is the best technique to find the best model.
