```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use("fivethirtyeight")
```

```python
def euler_vector(f,x,y0):
    n = np.shape(x)[0]
    # checking the input data
    ns = np.shape(y0)[0]
    # setting up a two column matrix to store the ans
    y = np.zeros((n,ns))
    dx = x[1] - x[0]
    y[0,:] = y0
    for i in range(n-1):
        y[i+1,:] = y[i,:] + dx * f(x[i],y[i,:])
    return y
```

```python
def problem(x,z):
    z0 = z[0]
    z1 = z[1]
    dz0 = z1
    dz1 = -z0 - 2*z1 + np.exp(-2*x)
    return np.array([dz0, dz1],dtype = float)
```

```python
x = np.linspace(0,5,100)
y0 = np.array([1.0, 2.5])
f = lambda x, z: np.array([z[1], -z[0]-2*z[1]+np.exp(-2*x)])
f2 = lambda x, z: np.array([z[1], -z[0]-2*z[1]+x**2])
z = euler_vector(f2, x, y0)
y = z[:,0]
fig = plt.figure(1, figsize = (14, 8)); plt.clf()
ax = fig.add_subplot(1,1,1)
ax.plot(z)
```
